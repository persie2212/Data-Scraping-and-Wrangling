## Naukri Website Datascraping for DataScience Jobs  

Datascraping is an essential tool to get data through vast resource of internet. It is mainly done through two methods: 
1)Using APIs
2)Scraping website
Here, we are scraping naukri website to get information on available data science jobs in India. This techniques uses extensive information of HTML tags and selenium library.

```python
#Importing Library
import pandas as pd
import bs4
import selenium
import requests
```

```python
#Importing webdriver, used to automatically run chrome
from selenium import webdriver
```

Naukri Website

```python
browser = webdriver.Chrome('chromedriver.exe')
naukri= "https://www.naukri.com/data-science-jobs"
browser.get(naukri)
```

```python
jobs = browser.find_elements_by_css_selector('div.row[type="tuple"]')
print(len(jobs))
```

```python
for job in jobs:
    job_title = job.find_element_by_css_selector('li.desig')
    job_title = job_title.get_attribute('title')
    print(job_title)
    
    company = job.find_element_by_css_selector('span.org')
    company = company.get_attribute('innerHTML')
    print(company)
```

```python
def get_info(job, selector, attribute):
    try:
        tag = job.find_element_by_css_selector(selector)
        tag = tag.get_attribute(attribute)
    except:
        tag = None
    return tag

df_jobs = pd.DataFrame(columns = ['title', 'company', 'salary', 'date', 
                                 'location', 'experience', 'description', 'url'])

for job in jobs:
    job_title = get_info(job, 'li.desig', 'title')
    company = get_info(job, 'span.org', 'innerHTML')
    salary = get_info(job, 'span.salary', 'innerHTML')
    date = get_info(job, 'span.date', 'innerHTML')
    
    parent_location = job.find_element_by_css_selector('span.loc')
    location = get_info(parent_location, 'span', 'innerHTML')
    experience = get_info(job, 'span.exp', 'innerHTML')
    description = get_info(job, 'span.desc', 'innerHTML')
    url = get_info(job, 'a.content','href' )
    
    curr_job = {'title': job_title,
               'company': company,
               'salary':salary,
               'date':date,
               'location':location,
               'experience':experience,
               'description':description,
               'url': url}
    df_jobs = df_jobs.append(curr_job, ignore_index = True)

df_jobs
```
